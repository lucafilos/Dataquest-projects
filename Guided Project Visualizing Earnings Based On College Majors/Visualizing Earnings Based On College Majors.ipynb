## Analysis of data distribution with histograms

```python
columns = ([
    'Sample_size',
    'Median',
    'Employed',
    'Full_time',
    'ShareWomen',
    'Unemployment_rate',
    'Total',
    'Men',
    'Women',
])

# optimal range
optimal_range=[]
for c in columns:
    describe=recent_grads.describe()
    Q1 = describe.loc['25%', c]
    Q3 = describe.loc['75%', c]
    IQR =  Q3- Q1
    if Q1 - 1.5*IQR < 0:
        minimum = 0
    else:
        minimum = Q1 - 1.5*IQR
            
    maximum = Q1 + 1.5*IQR
    optimal_range.append((minimum, maximum))
    
for c in range(len(columns)):
    fig = plt.figure(figsize = (10,5))
    plt.hist(recent_grads[columns[c]], range=(optimal_range[c]), bins=20)
    plt.title(columns[c])
    plt.show()
    
```

Hilights on histograms:
- Most of sample sizes are very low, this can question the accuracy of this study
- Salaries are around 35,000 USD/year
- Unemployment rate is around 4-8%

### Median income by major category

```python
recent_grads[['Median','Major_category']].boxplot(by='Major_category', figsize=(10,4), rot='vertical')
```

```python
grouped_data=recent_grads.groupby('Major_category').mean().sort_values('Median')
grouped_data.plot.barh(y='Median', figsize=(7,7))
```

Most of incomes ranges between 30,000 and 40,000 dollars.

`Business`, `Computers & Mathematics`, `Law & Public Policy` and `Physical Science` have higher incomes, between 40,000 and 50,000 dollars.

`Engineering` incomes are significantly higher, between 50,000 and 60,000 dollars but with a strong variance.

`Psychology & Social Work` have the least income range, less than 35,000 dollars


### Top 10 majors by income

```python
sorted_income=recent_grads.sort_values('Median', ascending=False)

sorted_income.head(10).plot.barh(x='Major', y='Median')
```

### Does number of students influences the median income?

Let's see if the number of students influences the median income. First I will divide the `Total` in 5 ranges.

```python
recent_grads.loc[recent_grads['Total'].between(0,1000),'Total_range']='0-1000'
recent_grads.loc[recent_grads['Total'].between(1000,10000),'Total_range']='1000-10k'
recent_grads.loc[recent_grads['Total'].between(10000,100000),'Total_range']='10k-100k'
recent_grads.loc[recent_grads['Total']>100000,'Total_range']='>100k'
```

```python
recent_grads[['Total_range','Median']].boxplot(by='Total_range', figsize=(10,4), rot='vertical')
```

We can see a little decrease in Median income as total number of students increases. It could be possible that majors with very few students tend to give more possibilities to find a better remunerated job because of a lower offer for those specialties in the market.

### Top 10 majors by unemployment rate
Let's see the top 10 majors by `Median` income and how they behave in unemployment rate

It's interesting to notice that **`Nuclear engineering`** leads to one of the best paid jobs but is the major with the highest unemployment rate. Also **`Mining and Mineral Engineering`** follows a similar pattern
